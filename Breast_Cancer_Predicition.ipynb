## Splitting the data into a training and a test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(breastcancer_dataset['data'], breastcancer_dataset['target'], random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import pandas as pd
iris_dataframe = pd.DataFrame(X_train, columns=breastcancer_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o', hist_kwds={'bins':20}, s=60, alpha=0.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=15)
```

```python
# Training a K-nearest neighbors classifier
knn.fit(X_train, y_train)
```

```python
# Evaluation of the classifier on the test set.
print("Accuracy on test data:", knn.score(X_test, y_test))
```
